```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

obj = list(df.select_dtypes(include='object').columns)

for i in obj:
  df[i] = encoder.fit_transform(df[i])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('Breakdown_or_Running_Late', axis=1)
y = df['Breakdown_or_Running_Late']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# i used simple imputer to handle missing value
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
imputer.fit_transform(x_train)
imputer.transform(x_test)
```

```python
# scale data by using Standard Scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Building the ANN Model
In this section, define the architecture of the ANN by specifying the number of layers, neurons, and activation functions.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
x_train_scaled.shape[1]
```

```python
# add input layer
model.add(Dense(units=6, input_dim= x_train_scaled.shape[1], activation='relu'))
```

```python
# add hidden layer and i assumed there is one hidden layer
model.add(Dense(units=6, activation='relu'))
```

```python
# add output layer with 1 unit because there is only one class
# let activation fuction been sigmoid because it works best with classification problems
model.add(Dense(units=1, activation='sigmoid'))
```

## Compile the Model
Compile the ANN model by defining the optimizer, loss function, and evaluation metrics.

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

## Training the Model
Train the ANN model using the training data.

```python
history = model.fit(x_train_scaled, y_train, batch_size=10, epochs=5, validation_split=0.2)
```

## Evaluate the Model
Evaluate the performance of the model on the test set.

```python
model.evaluate(x_test_scaled, y_test)
```

## Make Predictions
Use the trained model to make predictions on new or unseen data.

```python
y_pred = model.predict(x_test_scaled)
y_pred
```

## Model Performance Visualization
Visualize the performance metrics such as accuracy and loss over the epochs.

```python
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.legend()
plt.show()
```
